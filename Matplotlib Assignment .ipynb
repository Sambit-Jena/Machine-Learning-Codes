## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter Plot: It displays individual data points as markers on a 2D plane, where the position of each point represents its values for two variables. Scatter plots are helpful in examining relationships or clusters in the data.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y, c='#8F1342', alpha=.9, marker='+')
plt.xlabel(' Normal Dist. value of x')
plt.ylabel('Normal Dist. value of y')
plt.title( 'This is my x v/s y Scatter graph')
plt.grid()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.


The subplot() function in Matplotlib is used to create multiple subplots within a single figure. It allows you to arrange and organize multiple plots in a grid-like structure.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1,c='y')
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2,c='g')
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3,c='#980A6F')
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot 4')

plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of chart that represents categorical data using rectangular bars, where the length of each bar corresponds to the value of the category it represents. Bar plots are used to visualize and compare the values of different categories or to display the distribution of a single category.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit, color='#680034')
plt.grid()
plt.title("Company's profit in Billion $")
plt.xlabel('company')
plt.ylabel('profit')
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also called a box-and-whisker plot, shows the distribution of a dataset graphically. It presents the summary statistics in a condensed and graphic way, including the median, quartiles, and any outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.title('Box Plot - 1')
```

```python
plt.boxplot(box2)
plt.title('Box Plot - 2')
```
