## Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are visual markers used to represent data points or shapes in a plot. They are the fundamental building blocks for creating visualizations.Bokeh provides a wide range of glyphs, such as circles, squares, lines, rectangles, triangles, and more, which can be added to a plot to display data.

```python
from bokeh.sampledata.iris import flowers
output_file('test.html')

p=figure(title='test flower')
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

## Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

The Bokeh server is a key component of Bokeh that allows us to create and deploy interactive plots and applications that can be updated in real time. With the Bokeh server, we can build web-based applications that can handle user interactions, respond to events, and update the plots systematic and dynamic.

Use the Bokeh server to create interactive plots:
* 1- We can import any required module from the Bokeh and start polting the graph easily.
* 2- Define your plot by creating a figure object and adding glyphs to it, just like you would for a static plot.  We can define our plot by creating a figure object and adding glyphs to it.
* 3-  For including any interactive elements or arranging the plots we must specify the layout of our application by arranging the plots.
* 4- We can write callback functions that will be triggered in response to user interactions or events. These functions can update the data or properties of the plot dynamically.
* 5-  When we run the Bokeh server using the bokeh serve command in your terminal, followed by the filename of your Python script it helps to start the server and make our application accessible through a web browser.

## Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Basically we have to follow the the follow steps when ever we are trying to embed a Bokeh plot into a web page or dashboard using Flask and Django.

* step-1: First of all we need to create the Bokeh plot using the Bokeh library, as we normally do. This includes importing the necessary modules, defining the plot, adding glyphs, customizing the appearance, and configuring any interactivity.
* step-2: In the step-2 we supposed to use the "bokeh.embed" module to save the Bokeh plot as an HTML file that can be embedded into a web page. The "file_html()" function from bokeh.embed takes the plot object as an argument and returns the HTML code as a string.
* step-3:  In Flask, we can render a template that includes the Bokeh plot. First, create a Flask route that renders the template as we have very often.
 Then we can pass the plot_html variable containing the Bokeh plot HTML code to the template. After that we will create an HTML template (index.html) that includes a place holder for the Bokeh plot.
 But in case of Django, we can create a view function that renders a template with the Bokeh plot. Pass the plot_html variable containing the Bokeh plot HTML code to the template. And then proceeding with the same manner as Flask.

* step-4: In the very last step we start the Flask or Django server to run our web application. We can  navigate to the specified route or URL to view the web page with the embedded Bokeh plot.
 Aspecially for Flask we need to use the "flask run" command to execute and for Django "python manage.py" to run the server.
