### Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import pandas as pd
df = pd.DataFrame({'A': [10, 20, 30, 40],
                   'B': [50, 60, 70, 80],
                   'C': [90, 100, 110, 120]})
```

```python
df
```

```python
def reindex_dataframe(df):
    df.reset_index(drop=True, inplace=True)
    df.index = df.index.map(lambda x: x*2 + 1)
    return df
```

```python
df_reindexed = reindex_dataframe(df)
df_reindexed
```

### Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.

```python
import pandas as pd
df = pd.DataFrame({'Values': [10, 20, 30, 40, 10]})
```

```python
df
```

```python
def summation(df):
        addition = 0
        for index, row in df.iterrows():
            addition += row['Values']
        
            if index == 2:
                break

        print("Sum of the first three values=", addition)
```

```python
summation(df)
```

### Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
import pandas as pd
df = pd.DataFrame({'Text': ['sambit is a associate','kumar is my middle name','jena is my family name ','I am a aspiring data scientist']})
```

```python
df
```

```python
def count_words(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    return df
```

```python
count_words(df)
```

### Q5. How are DataFrame.size() and DataFrame.shape() different?

DataFrame.size:
    DataFrame.size returns the total number of elements in the DataFrame, which is equal to the number of rows multiplied by the number of columns. It gives the count of all elements, including NaN or missing values.

```python
# Example
import pandas as pd
df1 = pd.DataFrame({'A': [10, 20, 30, 40],
                   'B': [50, 60, 70, 80],
                   'C': [90, 100, 110, 120]})
```

```python
df1.size
```

DataFrame.shape:
DataFrame.shape returns a tuple that represents the dimensions of the DataFrame.The shape tuple contains two elements: the number of rows and the number of columns, respectively. And it  has  excluded the missing values.

### Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
import pandas as pd
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
df
```

```python
def statistics(df):
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_value = df['Values'].std()

    print("Mean=", mean_value)
    print("Median=", median_value)
    print("Standard Deviation=", std_value)
```

```python
statistics(df)
```

### Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
import pandas as pd

df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-07', '2023-01-08'],
                   'Sales': [10000, 22200, 1500, 300000, 123250, 3500, 400000, 45000]})

df['Date'] = pd.to_datetime(df['Date'])  # Convert the 'Date' column to datetime
```

```python
def MovingAverage(df):
    df = df.sort_values('Date')

    df['Moving Average'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df
```

```python
DF_MA = MovingAverage(df)
```

```python
DF_MA
```

### Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

```python
import pandas as pd


df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})
df
```

```python
def weekday(df):
    df['Date'] = pd.to_datetime(df['Date'])

    df['Weekday'] = df['Date'].dt.strftime('%A')

    return df
```

```python
WeekdayDataframe = weekday(df)
```

```python
WeekdayDataframe
```

### Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd
data= pd. date_range(start = '2022-12-21',end='2023-02-02')
df_date=pd.DataFrame(data)
```

```python
def rows_by_date(df):
 
    df[0] = pd.to_datetime(df[0])
    selected_rows = df[df[0].between('2023-01-01', '2023-01-31')]

    return selected_rows
```

```python
rows_by_date(df_date)
```
