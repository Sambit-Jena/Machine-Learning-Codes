## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating graphics. Here are five common plots that can be created using Seaborn along with their uses:

* Scatter Plot:

It shows the relationship between two continuous variables. It is useful for identifying patterns, clusters, or trends in the data. It can also be used for detecting outliers.

* Line Plot:

Line plot is use to displays the relationship between two continuous variables. It is commonly used to visualize trends over time, such as stock prices, temperature changes, or sales data.

* Bar Plot:

It represents categorical data with rectangular bars. It can be used to compare different categories or groups. It is often used to display counts, frequencies, or percentages.

* Pairplot:

The Pairplot allows us to plot pairwise relationships between variables within a dataset. This creates a nice visualisation and helps us understand the data by summarising a large amount of data in a single figure.

* Histogram:

It shows the distribution of a single numeric variable by dividing it into bins and displaying the frequency or count of values falling into each bin. It is useful for understanding the underlying distribution of the data.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
Fmri = sns.load_dataset('fmri')
```

```python
Fmri
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data = Fmri)
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
Titanic = sns.load_dataset("titanic")
```

```python
Titanic
```

```python
sns.boxplot(x='pclass', y='age', data= Titanic)
```

```python
sns.boxplot(x='pclass', y='fare', data = Titanic)
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
Diamonds = sns.load_dataset("diamonds")
```

```python
Diamonds
```

```python
sns.histplot(x='price', hue='cut', kde=True, data=Diamonds )
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
Flights = sns.load_dataset('flights')
```

```python
Flights
```

```python
Flights_data = Flights.pivot("month", "year", "passengers")
```

```python
sns.heatmap(data=Flights_data, annot=True, fmt="d", cmap="YlGnBu")
```
